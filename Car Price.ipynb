```python
numtype=['Mileage','Doors','Owner_Count','Price','Engine_Size']
```

```python
for col in numtype:
    print(f'Skewness of {col}:')
    print(df[col].skew())
```

```python
fig,axes=plt.subplots(3,2,figsize=(12,10))
axes=axes.ravel()

for i,col in enumerate(numtype):
    axes[i].hist(df[col],edgecolor='black',bins=20)
    axes[i].set_title(f'Frequncy Distribution of {col}')
    axes[i].tick_params(axis='x',rotation=50)
    axes[i].set_xlabel(f'{col}')
    axes[i].set_ylabel('Frequency')
plt.tight_layout()  
```

```python
fig,axes=plt.subplots(2,2,figsize=(12,10))
axes=axes.ravel()

for i,col in enumerate(chartype):
    sns.countplot(df,x=col,ax=axes[i],edgecolor='black',palette='Greens')
    axes[i].set_title(f'Frequncy Distribution of {col}')
    axes[i].tick_params(axis='x',rotation=80)
    axes[i].set_xlabel(f'{col}')
    axes[i].set_ylabel('Frequency')
plt.tight_layout();  
```

```python
pricemedian=df.groupby('Brand',as_index=False)['Price'].median()
pricemedian
```

```python
pricemedian=df.groupby('Brand',as_index=False)['Price'].median()

plt.figure(figsize=(12,6))
plt.barh(pricemedian['Brand'],pricemedian['Price'],edgecolor='black',color=['red','blue','green','orange','purple'])
plt.title('Average Price of different Car Brands(2002-2003)')
plt.xlabel('Average Price')
plt.ylabel('Brands');
```

```python
mileagemean=df.groupby('Brand',as_index=False)['Mileage'].mean()

plt.figure(figsize=(12,6))
plt.barh(mileagemean['Brand'],mileagemean['Mileage'],edgecolor='black',color=['red','blue','green','orange','purple'])
plt.title('Average Mileage of different Car Brands(2002-2023)')
plt.xlabel('Average Mileage')
plt.ylabel('Brands');
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(df,x='Price',y='Engine_Size',hue="Price")
plt.title('Engine Size Vs Price')
plt.xlabel('Price')
plt.ylabel('Engine Size');
```
